```python
socialNetwork = pd.read_csv('Social.csv')
```

```python
socialNetwork.head()
```

```python
X = socialNetwork.iloc[:,:-1].values
X
```

```python
Y = socialNetwork.iloc[:,-1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
#Feature scaling - should be done after the data is split into test and training set
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
```

```python
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
```

```python

cVal =np.arange(1,100)
max_score = 0
best_cVal = 0
scores = []

for cval in cVal:
    logisticRegressor = LogisticRegression(C=cval)
    logisticRegressor.fit(X_train,Y_train)
    score = logisticRegressor.score(X_test,Y_test)
    scores.append(score)
    print(score)
    print(cval)
    
    if max_score < score:
        max_score = score
        best_cVal = cval
        
        
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(cVal,scores)
ax.set_yscale('log')
ax.set_xlabel('No of trees')
ax.set_ylabel('R2 score')
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegressor = LogisticRegression()
```

```python
logisticRegressor.fit(X_train,Y_train)
```

```python
logisticRegressor.score(X_test,Y_test)
```

```python
y_pred = logisticRegressor.predict(X_test)
```

```python
# predicting a new result
print(logisticRegressor.predict(sc_x.fit_transform([[30,87000]])))
```

```python
print(np.column_stack((Y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(Y_test,y_pred)
print(cm)
accuracy_score(Y_test,y_pred)
```

```python
sns.heatmap(cm,annot=True)
```

```python
# Visualizing the training set results 
from matplotlib.colors import ListedColormap
X_set, y_set = sc_x.inverse_transform(X_train), Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, logisticRegressor.predict(sc_x.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
#Visualizing the test set result
from matplotlib.colors import ListedColormap
X_set, y_set = sc_x.inverse_transform(X_test), Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, logisticRegressor.predict(sc_x.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
